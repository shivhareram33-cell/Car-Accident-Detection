```python
import os
import random
import matplotlib.pyplot as plt

# Paths
image_dir = "../custom_dataset/images/train"
label_dir = "../custom_dataset/labels/train"

# Class names from data.yaml
class_names = ['Accident']

# Get image files
image_files = [f for f in os.listdir(image_dir)
               if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

# Randomly select 5 images
selected_images = random.sample(image_files, 5)

# Show images with labels
for img in selected_images:
    plt.figure(figsize=(5, 5))
    plt.imshow(plt.imread(os.path.join(image_dir, img)))

    # Find corresponding label file
    label_file = os.path.splitext(img)[0] + ".txt"
    label_path = os.path.join(label_dir, label_file)

    if os.path.exists(label_path):
        with open(label_path, "r") as f:
            labels = f.readlines()

        # Get unique class names
        label_names = set()
        for line in labels:
            class_id = int(line.split()[0])
            if class_id < len(class_names):
                label_names.add(class_names[class_id])
            else:
                label_names.add(f"Class {class_id}")

        title = ", ".join(label_names)
    else:
        title = "No label"

    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
import os

label_dirs = [
    "../custom_dataset/labels/train",
    "../custom_dataset/labels/val"
]

class_ids = set()

for label_dir in label_dirs:
    for label_file in os.listdir(label_dir):
        if label_file.endswith(".txt"):
            with open(os.path.join(label_dir, label_file), "r") as f:
                for line in f:
                    class_id = int(line.split()[0])
                    class_ids.add(class_id)

print("Unique class IDs:", class_ids)
print("Number of classes:", len(class_ids))
```
